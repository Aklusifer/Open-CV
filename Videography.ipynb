### Connecting a webcam

```python
import cv2
import time
import numpy as np
```

```python
cap = cv2.VideoCapture(0)         #for webcam

time.sleep(2)
width =int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height =int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(width, height)

while True:
    
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    #-------------------------------------------------------
    print('ret',ret)
    
    if ret == False:
        break
        
    #Display the resulting frame
    cv2.imshow('frame',frame)
    #-------------------------------------------------------
    
    #this command let's us quit with the 'q' button on a keyboard.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

### Create custom functions to rotate images in any  point(default is center)

```python
def rotate(img,angel, center=None,scale=1.0):
    
    (h,w)= img.shape[:2]
    
    if center==None:
        center=(w/2,h/2)
        
#create matrix for rotation
    M=cv2.getRotationMatrix2D(center,angel,scale)
    rotated= cv2.warpAffine(img,M,(w,h))
    
    return rotated
```

```python
#rotate the image by 120 degree in center point
rot = rotate(img,+60)
cv2.imshow('Rotated image',rot)
cv2.waitKey()
cv2.destroyAllWindows()
```
